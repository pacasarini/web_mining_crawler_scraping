```python
from sklearn.feature_extraction.text import TfidfVectorizer

stopwords = nltk.corpus.stopwords.words('portuguese')

import nltk
stemmer = nltk.stem.RSLPStemmer()

def stemmer_tokenizer(doc):
  # considerar somente termos com letras e pelo menos 3 caracteres
    termos = [termo.lower() for termo in nltk.word_tokenize(doc)
              if re.match(r'^[^\W\d]{3,}$', termo)]
    # aplicar stemmer
    termos = [stemmer.stem(termo) for termo in termos]
    return termos

vectorizer = TfidfVectorizer(
    binary=False,
    norm='l1',
    use_idf=True,
    stop_words=stopwords,
    tokenizer=stemmer_tokenizer
)
vetor = vectorizer.fit_transform([texto1, texto2])
print(vetor.toarray())
```

Aplicação para um conjunto de textos

```python
import pandas as pd
avaliacoes = pd.read_excel('/content/drive/My Drive/Colab Notebooks/nlp/aula/data/kindle_review/reviews.xlsx')
```

```python
avaliacoes
```

```python
vectorizer = TfidfVectorizer(
    binary=False,
    norm='l1',
    use_idf=True,
    stop_words=stopwords,
    tokenizer=stemmer_tokenizer
)
vetor = vectorizer.fit_transform(avaliacoes.texto)
print(vetor)
```

```python
X = vetor
y = avaliacoes.categoria
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from sklearn import tree
from sklearn import neural_network
from sklearn.metrics import accuracy_score

# inicar modelo
dtree = tree.DecisionTreeClassifier(random_state=88)
#treinar modelo
dtree.fit(X_train, y_train)

# classificar conjunto de testes
y_predict = dtree.predict(X_test)

# avaliar o modelo acucárica
print(accuracy_score(y_predict, y_test))

# inicar modelo
dtree = neural_network.MLPClassifier(
    hidden_layer_sizes=(10,),
    activation='logistic',
    solver='sgd',
    learning_rate='constant',
    learning_rate_init=0.05,random_state=88 )
#treinar modelo
dtree.fit(X_train, y_train)

# classificar conjunto de testes
y_predict = dtree.predict(X_test)

# avaliar o modelo acucárica
print(accuracy_score(y_predict, y_test))

```
